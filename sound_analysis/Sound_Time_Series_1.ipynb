##### 1.	Create 3 sine waves with a sample rate of 8000 Hz, 2 seconds each in length with frequencies and amplitudes:     * 110Hz at 100% amplitude     * 220Hz at 75% amplitude     * 440Hz at 50% amplitude

```python
fs = 8000
dur = 2
freq = [110, 220, 440]
A = [1, .75, .5]
sample_len = 2 * fs 
sines = [[],[],[]]
T = np.linspace(0, dur, num=sample_len)

for j in range(3):
    for i in range(sample_len):
        sines[j].append(A[j] * np.sin(2 * np.pi * freq[j] * T[i]))
```

8.	Create an algorithm that detects the combined signals spectral peaks and returns their frequencies in Hz (I will be arbitrarily varying the amplitude and frequency parameters of your previously created signals to test this)

```python
def detection(com_magnitude_spectrum):
    thresh = np.mean(com_magnitude_spectrum) + (4 * np.std(com_magnitude_spectrum))    
    com_mag_spec_thr = (com_magnitude_spectrum > thresh) * com_magnitude_spectrum
    
    diff_spec = np.diff(com_mag_spec_thr)
    diff_spec = np.roll(diff_spec,1)
    match = np.convolve(np.sign(diff_spec), [-1,1])
    match = np.roll(match, 1)
    idx = np.nonzero(match>0)[0]-2
    
    for i, peak in enumerate(idx):
        print "Element %s is a peak at frequency %s" % (idx[i], freqs[idx][i])
              
```
