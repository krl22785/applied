##### Part 1: Load the urban audio 

```python
fs, samples = wavfile.read('audio/brooklyn_street.wav')
```

```python
dur = float(len(samples)) / fs
sample_len = len(samples)
```

```python
samples = samples / 32768.0
T = np.linspace(0, dur, num = sample_len)
ab_samples = np.abs(samples)
```

###### Part 3: Implement Moving Average

```python
## can use different window size 
window_size = int(.125*fs) ##5512 
#window_size = 8192

hop_size = window_size/2
w = np.hanning(window_size)

window_avg = [np.sqrt((np.sum([i ** 2 for i in ab_samples[i:i+window_size]])/window_size)) \
                for i in range(0, len(ab_samples), hop_size) if i+window_size <= len(ab_samples)]

T = np.linspace(0, dur, num = len(window_avg))
```

##### Part 4: Plot this moving RMS array on the plot generated in part 2

```python
T = np.linspace(0, dur, num = sample_len) ##change T for original waveform 
fig, ax = plt.subplots(1,1, figsize=(10,4))
ax.plot(T, samples, alpha = .25)
ax.set_title("Brooklyn Sound")
ax.set_xlabel("Time Domaim (s)")
ax.set_ylabel("Amplitude")

T = np.linspace(0, dur, num = len(window_avg)) ##change T for moving average 
ax.plot(T, window_avg, color = 'r')
plt.xlim([0, np.max(T)])
plt.title('Brooklyn street')
plt.xlabel('Time (s)')
plt.ylabel('Absolute amplitude')
plt.grid('on')
```

##### Part 7: Detection Algo

7.  Implement an event detection algorithm that will automatically identify salient sound sources within the urban audio recording. Salient sound sources mean sounds that stand out from the background noise as distinct sonic events.
    * The algorithm can use a combination of time domain and frequency domain techniques
    * The algorithm should utilise a moving window approach
    * Add individually colored regions (<1.0 alpha) to a time domain and frequency domain (spectrogram) plots identifying the start and end time of each sound event
    * Output start time, end time, duration, amplitude (RMS) and at least two spectral features for each identified event
    * Save each identified sound event as a wave file, named by its start time in milliseconds
    * There are around 10 salient events in the urban audio recording

```python
dft_output = np.fft.rfft(w * samples[0:window_size])
magnitude_spectrum = [np.sqrt(i.real**2 + i.imag**2)/len(dft_output) for i in dft_output]
freqs = np.linspace(0,fs/2, num=len(dft_output))
```

```python
mag_spec_array_complex = [np.fft.rfft(w * samples[i:i + window_size]) \
                          for i in range(0, len(samples), hop_size)if i + window_size <= len(samples)]

mag_spec_array = []

for x in range(0, len(mag_spec_array_complex)):
    mag_spec_array.append([np.sqrt(i.real**2 + i.imag**2)/len(mag_spec_array_complex[x]) for i in mag_spec_array_complex[x]])

spec_cent_array = np.zeros(len(mag_spec_array))

for x in range(0, len(mag_spec_array)):
    spec_cent_array[x] = np.sum(mag_spec_array[x]*freqs) / np.sum(mag_spec_array[x])

T = np.linspace(0, dur, num=len(window_avg))

def centralSpectroid(spec_cent_array):
    plt.plot(T,spec_cent_array, alpha = .5)
    plt.xlim([0, np.max(T)])
    plt.title('Brooklyn street - spectral centroid over time')
    plt.ylabel('Spectral centroid (Hz)')
    plt.grid('on')
    
def centralSpectroidDifferences(spec_cent_array):
    plt.plot(T[0:len(np.diff(spec_cent_array))], np.diff(spec_cent_array), alpha = .5)
    plt.xlim([0, np.max(T)])
    plt.title('Brooklyn street - spectral centroid differences over time')
    plt.ylabel('Spectral centroid differences (Hz)')
    plt.grid('on')
    
def spectralFlatness(spec_cent_array):
    spec_flat_array = np.zeros(len(mag_spec_array))

    for x in range(0, len(mag_spec_array)):
        spec_flat_array[x] = sps.mstats.gmean(mag_spec_array[x]) / np.mean(mag_spec_array[x]) 
        
    plt.plot(T,spec_flat_array, alpha = .5)

    plt.xlim([0, np.max(T)])
    plt.title('Brooklyn street - spectral flatness over time')
    plt.xlabel('Time (s)')
    plt.ylabel('Spectral flatness')
    plt.grid('on')
```

```python
def urbanThreshold(spec_cent_array):
    T = np.linspace(0, dur, num=len(window_avg))
    thresh_abs = np.std(np.abs(np.diff(spec_cent_array))) * 2.5 

    urbanSounds = (np.diff(spec_cent_array) > thresh_abs) * np.diff(spec_cent_array)
    v = np.vstack((urbanSounds,T[0:len(np.diff(spec_cent_array))]))
    indx = np.nonzero(v[0])
    newArray = zip(v[1][indx][::2],v[1][indx][1::2])
    
    convert = len(samples)/44100.00
    
    for cnt, i in enumerate(newArray):
        startTime = i[0]
        endTime = i[1]
        timeRange = round(i[1]-i[0],2)
        print "Sound %s - start time: %s, end time: %s, Duration: %s secs" % (cnt, startTime, endTime, timeRange)
        
        samplesToSave = samples[int(startTime*44100):int(endTime*44100)]
        name = str(startTime) + '.wav'
        
        samplesToSave = np.int16(samplesToSave * 32768)
        wavfile.write('audio/'+name, fs, samplesToSave)
        

    plt.plot(T,spec_cent_array, alpha = .5)
    plt.xlim([0, np.max(T)])
    plt.title('Brooklyn street - Urban Sound Detection')
    plt.ylabel('Spectral centroid (Hz)')
    plt.grid('on')

    for i in newArray:
        plt.axvspan(i[0], i[1], facecolor = 'green', alpha = .25)
```
