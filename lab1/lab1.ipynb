```python
# from eddy 
from sklearn import svm
import pandas as pd
import scipy as sp
import numpy as np
import statsmodels.api as sm # 
import statsmodels.formula.api as smf #
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn import datasets 
from patsy import dmatrices

#load data 
fileUrl = 'https://raw.githubusercontent.com/EricSchles/MultipleLinearRegression/master/trafficking_data.csv'
trafficData = pd.read_csv(fileUrl)

adultVictims = smf.ols('trafficData["Adult victims"] ~ trafficData["persons prosecuted"] + trafficData["gdp"] + trafficData["policy index"] + trafficData["% females in primary education"] + trafficData["life expectancy"]', data=trafficData).fit()
childVictims = smf.ols('trafficData["child victims"] ~ trafficData["persons prosecuted"] + trafficData["gdp"] + trafficData["policy index"] + trafficData["% females in primary education"] + trafficData["life expectancy"]', data=trafficData).fit()

y_persons_prosecuted = smf.ols('trafficData["Adult victims"] ~ trafficData["persons prosecuted"]', data=trafficData).fit()
y_gdp = smf.ols('trafficData["Adult victims"] ~ trafficData["gdp"]', data=trafficData).fit()
y_policy_index = smf.ols('trafficData["Adult victims"] ~ trafficData["policy index"]', data=trafficData).fit()
y_education = smf.ols('trafficData["Adult victims"] ~ trafficData["% females in primary education"]', data=trafficData).fit()
y_life = smf.ols('trafficData["Adult victims"] ~ trafficData["life expectancy"]', data=trafficData).fit()


print "R-Squared for Independent Variables"
print "Females in Primary Education: %s" % round(y_education.rsquared,4)
print "Policy Index: %s" % round(y_policy_index.rsquared,4)
print "Life Expectance: %s" % round(y_life.rsquared,4)
print "Persons Prosecuted: %s" % round(y_persons_prosecuted.rsquared,4)
print "GDP R-Squared: %s" % round(y_gdp.rsquared,4)

print adultVictims.summary()
print '\n'
print childVictims.summary()
```

# The variable with the most explanatory power is the 'percent of females in primary education'.  We determine this by  comparing the R-Squared for each bivariate model, where adult victims is Y.  Again, R-Squared don't have any significant  meaning unless it is being used to compared relative models.  If we run a mutli-variate model, and compare the relative coefficient values, we can confirm that any changes in the value of 'percent of females with primary education' will impact the Y-value the most.    All of the above holds true if we make child victims Y.

```python
from matplotlib.colors import ListedColormap
# Create color maps for 3-class classification problem, as with iris
cmap_light = ListedColormap(['#FFAAAA', '#AAFFAA', '#AAAAFF'])
cmap_bold = ListedColormap(['#FF0000', '#00FF00', '#0000FF'])

def plot_estimator(estimator, X, y):
    
    try:
        X, y = X.values, y.values
    except AttributeError:
        pass
    
    estimator.fit(X, y)
    x_min, x_max = X[:, 0].min() - .1, X[:, 0].max() + .1
    y_min, y_max = X[:, 1].min() - .1, X[:, 1].max() + .1
    xx, yy = np.meshgrid(np.linspace(x_min, x_max, 100),
                         np.linspace(y_min, y_max, 100))
    Z = estimator.predict(np.c_[xx.ravel(), yy.ravel()])

    # Put the result into a color plot
    Z = Z.reshape(xx.shape)
    plt.figure()
    plt.pcolormesh(xx, yy, Z, cmap=cmap_light)

    # Plot also the training points
    plt.scatter(X[:, 0], X[:, 1], c=y, cmap=cmap_bold)
    plt.axis('tight')
    plt.axis('off')
    plt.tight_layout()
    
    plot_estimator(svc, X, y)
```
