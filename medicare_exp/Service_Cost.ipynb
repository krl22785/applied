###### What is Medicare?

Medicare is the social insurance program designed to pay for health services for the elderly and the dsiabled.  The program serves more than 44 million enrollees as of 2008 and cost the U.S. approximately $432 Billion, or 3.2% of GDP in 2007. The Medicare program consists of two parts for hospitals and medical insurance (Part A) and (Part B) and two additional services that provide flexibility and prescription drugs (Part C and Part D).  
    
Part A, hospital insurance (HI) covers all expenses incurred during a hospital stay.  This includes meals, supplies, testing, and a semi-private room. HI also covers home health care, medical equipment and walkers and wheelchairs.  Part B, supplementary medical insurance (SMI) pays for medically necessary physician visits, outpatient visits, home health care costs and other services for the elderly and disabled. 

As part of the Obama Administration's effort to make healthcare more transparent, affordable and accountable, the Center's for Medicare and Medicaid Services (CMS) published a dataset titled the Provider Utilization and Payment Data Inpatient Public Use File (Inpatient PUF).  The dataset contains information on services and and procedures provided to holders of Medicare by hospitals nationwide.  

The Inpatient PUF contains incurred charges by more than 3,000 U.S. hospitals that accept Medicare Inpatient Propsective Payment System (IPPS) payments for the top 100 most billed discharges.  The dataset contains the rate billed per discharge using the Medicare Severity Diagnosis Related Group (MS-DRG).  In order to clarify, the Diagnosis Related Group is a system to clasify hospital stays into one of about 500 groups, with the last group being labeled "Ungroupable".  The intent of the classification system was to identify the products and services that a hopsital provides based on diagnosis.  DRGs have been used in the U.S. since 1982 as a heuristic to gauge how much Medicare pays a hospital for each product and services incurred, sinced patients in a category are clincially similar.  

```python
df = pd.read_csv('IPPS.csv')
df.columns = ['drg_definition', 'provider_id', 'provider_name', 'address', 'city', 'state', 'provider_zip_code', 'hospital_referral_region_desc', 'total_discharges', 'average_covered_charges','average_total_payments','average_medicare_payments']

df['average_covered_charges'] = df['average_covered_charges'].str.replace('$','').astype(float)
df['average_total_payments'] = df['average_total_payments'].str.replace('$','').astype(float)
df['average_medicare_payments'] = df['average_medicare_payments'].str.replace('$','').astype(float)

df['out_pocket']= df['average_total_payments'] - df['average_medicare_payments']
df['unpaid_balance'] = df['average_covered_charges'] - df['average_total_payments']

header = list(df.columns.values)
```

The acquired dataset was retrieved from CMS's Medicare Provider Analysis and Review (MDPAR) inpatient data for FY2011.  The dataset contains 163065 observations and some contains the following attributes: 

* DRG The code and desciption identifying the MS-DRG.  DRGs are classification systems for identifying products used or services incurred per diagnosis 

* Provider
    * Name 
    * Street
    * City
    * State
    * Zip Code 
* Region
* Total Discharges which is a total number of patients to incur services 
* Average Covered Charges which is the charge for services covered by Medicare 
* Average Total is payments to all providers for the MS-DRG including the MS-DRG amount, teaching, capital, and outlier payments for all cases 
* Average Medicare Payments is the average amount that Medicare pays to the provider for Medicare's share of the MS-DRG

### Total Medical Cost Delinquencies

We will evalute if the average cost of services drives a high number of medical payment delinquencies per data from the U.S. Consumer Financial Protection Bureau (CFPB).  

* 194 - SIMPLE PNEUMONIA & PLEURISY W CC
* 690 - KIDNEY & URINARY TRACT INFECTIONS W/O MCC
* 292 - HEART FAILURE & SHOCK W CC
* 392 - ESOPHAGITIS, GASTROENT & MISC DIGEST DISORDERS W/O MCC
* 641 - MISC DISORDERS OF NUTRITION,METABOLISM,FLUIDS/ELECTROLYTES W/O MCC


#### By State 

```python
df_ = pd.read_csv('Consumer_Complaints.csv', index_col='Date received')
finance = df_[df_["Sub-product"] == "Medical"][["State", "Issue", "Sub-issue"]]

grps = finance.groupby('State')
delinquencies = grps.count()
del delinquencies['Issue']
del delinquencies['Sub-issue']
delinquencies.columns = ['Count']
delinquencies 
delinquencies.index.names = ['state']
```

#### By Zip 

```python
finance_zip = df_[df_["Sub-product"] == "Medical"][["ZIP code", "Issue", "Sub-issue"]]
```

```python
services = ['194 - SIMPLE PNEUMONIA & PLEURISY W CC', '690 - KIDNEY & URINARY TRACT INFECTIONS W/O MCC', \
    '292 - HEART FAILURE & SHOCK W CC', '392 - ESOPHAGITIS, GASTROENT & MISC DIGEST DISORDERS W/O MCC', \
    '641 - MISC DISORDERS OF NUTRITION,METABOLISM,FLUIDS/ELECTROLYTES W/O MCC']
    
    
stateCharge = df[df['drg_definition'].isin(services)][["state", "drg_definition", "average_covered_charges"]]
```

```python
stateCharge['average_covered_charges'] = stateCharge['average_covered_charges'].str.replace('$', '').astype(float)
grouped = stateCharge.groupby(['state', 'drg_definition'])
stateChargeMean = grouped.mean()
```

```python
stateChargeMean1 = stateChargeMean.reset_index().set_index('state')
pneu = stateChargeMean1[stateChargeMean1['drg_definition'] == '194 - SIMPLE PNEUMONIA & PLEURISY W CC']
#merge tables on columns, not index 
delinquencies['state'] = delinquencies.index
pneu['state'] = pneu.index
nm = pd.merge(pneu, delinquencies, on='state')


kidney = stateChargeMean1[stateChargeMean1['drg_definition'] == '690 - KIDNEY & URINARY TRACT INFECTIONS W/O MCC']
delinquencies['state'] = delinquencies.index
kidney['state'] = kidney.index
km = pd.merge(kidney, delinquencies, on='state')
```

```python
plt.figure(figsize=(16,6))

pneuModel = smf.ols('Count ~ average_covered_charges', data = nm).fit()
print "Coefficient Value: %s" % pneuModel.params[1]

plt.subplot(121)
plt.scatter(nm['average_covered_charges'], nm['Count'], \
            color = 'r', alpha = .6, s = 50)
plt.xlabel("Average Covered Costs \n(Pneumonia)")
plt.ylabel("CFPB Complaints for Medical Bills")
plt.text(10000, 550, r'coefficient = %s' % pneuModel.params[1], fontsize=12)
plt.title('FIGURE 3 \n\n %s' % services[0])


kidneyModel = smf.ols('Count ~ average_covered_charges', data = km).fit()
print kidneyModel.params[1]

plt.subplot(122)
plt.scatter(km['average_covered_charges'], km['Count'], \
            color = 'r', alpha = .6, s = 50)
plt.xlabel("Average Covered Costs \n(Kidney)")
plt.ylabel("CFPB Complaints for Medical Bills")
plt.text(10000, 550, r'coefficient = %s' % kidneyModel.params[1], fontsize=12)
plt.title('FIGURE 4 \n\n %s' % services[1])
```

For this exercise we looked at the most popular DRG, simple pneumonia and pleurisy.  We modeled the relationship between the total number of individuals to complain about medical collection agency calls as a function of the average covered charges.  Interestingly, coefficient value is close to zero.  An increase in incurred cost does not result in more users complaining to the CFPB over collection agencies calling over medical expenses. 

This is interesting because one of the objectives of Medicare is to balance out the health care costs based a variety of external variables and conditions.  Essentially, pricing for incurred services is adjust based on market conditions and the area's wage index.
